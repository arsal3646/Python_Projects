## SECTION 6: Type Conversion

Type conversion means changing a value from one type to another.

Common conversion functions:

- `int()`
- `float()`
- `str()`

### Why conversion is needed

Data often comes from files as text.

For example:

```python
"1000"
```

looks like a number, but because it is inside quotes, Python treats it as text.

To calculate with it, convert it.

````python
amount_text = "1000"
amount_number = int(amount_text)

print(amount_number + 500)
````

### Converting number to string

If you want to join text and a number using `+`, convert the number to a string.

````python
salary = 15000
message = "My salary is " + str(salary)

print(message)
````

### Common mistake: adding string and number

The following code is broken:

```python
salary = 15000
message = "My salary is " + salary
```

Why?

Because `"My salary is "` is a string, but `salary` is an integer.

Python does not automatically decide how to combine them.

````python
# Broken example shown safely using try/except

try:
    salary = 15000
    message = "My salary is " + salary
    print(message)
except TypeError as error:
    print("TypeError happened:")
    print(error)
````

### Better method: f-string

An f-string is usually cleaner.

````python
salary = 15000
message = f"My salary is {salary}"

print(message)
````

## SECTION 14: Functions

A function is a reusable block of code.

### Why functions exist

Without functions, you repeat the same logic again and again.

With functions, you define logic once and reuse it.

### Basic structure

```python
def function_name(parameter_1, parameter_2):
    result = parameter_1 + parameter_2
    return result
```

Important words:

- `def` starts a function definition.
- Parameters are inputs.
- `return` sends a value back.

````python
def calculate_profit(revenue, cost):
    profit = revenue - cost
    return profit

result = calculate_profit(50000, 35000)

print(result)
````

### Profit margin function

Profit margin is:

```text
profit / revenue
```

Example: if profit is 15000 and revenue is 50000, margin is 0.30 = 30%.

````python
def calculate_profit(revenue, cost):
    profit = revenue - cost
    return profit

def calculate_margin(profit, revenue):
    margin = profit / revenue
    return margin

profit = calculate_profit(50000, 35000)
margin = calculate_margin(profit, 50000)

print(margin)
print(f"Profit margin is {margin:.2%}")
````

### Difference between `print` and `return`

`print()` shows something on the screen.

`return` gives a value back to the program so it can be used later.

Most useful functions should return values.

````python
def add_numbers(a, b):
    return a + b

answer = add_numbers(10, 20)

print("Answer:", answer)
````

## SECTION 15: Common Errors

Errors are not failure. Errors are diagnostic messages.

**Diagnostic** means helping identify the problem.  
Urdu anchor: **tashkheesi / تشخیصی**.

Common Python errors:

- `SyntaxError`
- `NameError`
- `TypeError`
- `ZeroDivisionError`
- `IndentationError`

We will look at each error safely.

### 15.1 SyntaxError

A syntax error means Python grammar is broken.

Broken code:

```python
print("Hello"
```

Correct code:

```python
print("Hello")
```

````python
# Safe demonstration of SyntaxError using exec()

broken_code = 'print("Hello"'

try:
    exec(broken_code)
except SyntaxError as error:
    print("SyntaxError happened:")
    print(error)

print("Correct version:")
print("Hello")
````

### 15.2 NameError

A name error happens when you use a variable name that does not exist yet.

Broken code:

```python
print(total_revenue)
```

Correct code:

```python
total_revenue = 50000
print(total_revenue)
```

````python
try:
    print(total_revenue)
except NameError as error:
    print("NameError happened:")
    print(error)

total_revenue = 50000
print("Correct total revenue:", total_revenue)
````

project_names = ["Tower A", "Villa B", "Mall C"]

Tasks:

   - On Budget if actual == budget

### Version 1: Direct approach without functions

Predict the output before running.

````python
project_names = ["Tower A", "Villa B", "Mall C"]
budgets = [100000, 150000, 200000]
actuals = [120000, 140000, 200000]

for index in range(len(project_names)):
    project = project_names[index]
    budget = budgets[index]
    actual = actuals[index]
    variance = actual - budget

    if actual > budget:
        status = "Over Budget"
    elif actual < budget:
        status = "Under Budget"
    else:
        status = "On Budget"

    print("Project:", project)
    print("Budget:", budget)
    print("Actual:", actual)
    print("Variance:", variance)
    print("Status:", status)
    print("-" * 30)
````

### Version 2: Improved approach with functions

Now we make the logic reusable.

Function 1:

```python
calculate_variance(budget, actual)
```

Function 2:

```python
classify_project(budget, actual)
```

````python
def calculate_variance(budget, actual):
    return actual - budget

def classify_project(budget, actual):
    if actual > budget:
        return "Over Budget"
    elif actual < budget:
        return "Under Budget"
    else:
        return "On Budget"

project_names = ["Tower A", "Villa B", "Mall C"]
budgets = [100000, 150000, 200000]
actuals = [120000, 140000, 200000]

for index in range(len(project_names)):
    project = project_names[index]
    budget = budgets[index]
    actual = actuals[index]

    variance = calculate_variance(budget, actual)
    status = classify_project(budget, actual)

    print("Project:", project)
    print("Budget:", budget)
    print("Actual:", actual)
    print("Variance:", variance)
    print("Status:", status)
    print("-" * 30)
````

### Final challenge: overall summary

Calculate:

- total budget,
- total actual cost,
- total variance,
- overall status.

````python
def calculate_variance(budget, actual):
    return actual - budget

def classify_project(budget, actual):
    if actual > budget:
        return "Over Budget"
    elif actual < budget:
        return "Under Budget"
    else:
        return "On Budget"

project_names = ["Tower A", "Villa B", "Mall C"]
budgets = [100000, 150000, 200000]
actuals = [120000, 140000, 200000]

total_budget = sum(budgets)
total_actual = sum(actuals)
total_variance = calculate_variance(total_budget, total_actual)
overall_status = classify_project(total_budget, total_actual)

print("OVERALL SUMMARY")
print("-" * 30)
print("Total budget:", total_budget)
print("Total actual:", total_actual)
print("Total variance:", total_variance)
print("Overall status:", overall_status)
````
