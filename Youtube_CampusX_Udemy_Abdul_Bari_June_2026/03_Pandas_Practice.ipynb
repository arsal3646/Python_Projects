## Section 5 — Selecting rows with `loc` and `iloc`

Pandas gives two main tools for selecting rows:

```text
loc  = label-based selection
iloc = position-based selection
```

First-principles distinction:

```text
iloc asks: where is the row by number position?
loc asks: what is the row label?
```

For now, because our index is 0, 1, 2, 3, both may look similar. Later, when index labels change, the difference matters.


````python
df
````

````python
# Select first row by position

df.iloc[0]
````

````python
# Select first three rows by position

df.iloc[0:3]
````

````python
# Select row with index label 0

df.loc[0]
````

````python
# Select rows 0 to 2 using loc

df.loc[0:2]
````

### Important difference in slicing

With `iloc`:

```python
df.iloc[0:3]
```

includes positions 0, 1, 2, but excludes 3.

With `loc`:

```python
df.loc[0:2]
```

includes labels 0, 1, 2.

For numeric default index, this can feel confusing. Go slowly.


### Selecting rows and columns together

With `loc`:

```python
df.loc[rows, columns]
```


````python
df.loc[0:2, ["Project", "Budget"]]
````

````python
df.iloc[0:3, 0:3]
````

### Multiple conditions

For multiple conditions, use:

```python
&   means AND
|   means OR
```

Each condition must be inside parentheses.

Example:

```python
df[(df["Actual"] > df["Budget"]) & (df["Revenue"] > 200000)]
```

This is wrong:


Why?

Because Pandas needs parentheses around each condition.

Correct:

```python
df[(df["Actual"] > df["Budget"]) & (df["Revenue"] > 200000)]
```

## Section 8 — Creating status columns using conditions

Business data often needs classification.

Examples:

- Over Budget
- Under Budget
- On Budget

We can use `.apply()` with a function, or use NumPy. Since this is Pandas practice, let us first use a simple function.


````python
def classify_budget(row):
    if row["Actual"] > row["Budget"]:
        return "Over Budget"
    elif row["Actual"] < row["Budget"]:
        return "Under Budget"
    else:
        return "On Budget"

df["Budget_Status"] = df.apply(classify_budget, axis=1)

df
````

### What does `axis=1` mean?

When using `.apply()`:

```text
axis=0 means apply down columns
axis=1 means apply across rows
```

Here, each row contains Budget and Actual for one project, so we use:

```python
axis=1
```

## Section 10 — Missing values

Real datasets are often incomplete.

Missing values in Pandas appear as:

```text
NaN
```

NaN means:

```text
Not a Number
```

Missing data matters because it can distort totals, averages, models, and conclusions.


````python
data_with_missing = pd.DataFrame({
    "Project": ["Tower A", "Villa B", "Mall C", "Road D"],
    "Budget": [100000, 150000, None, 175000],
    "Actual": [120000, None, 210000, 190000],
    "Revenue": [160000, 180000, 260000, None]
})

data_with_missing
````

````python
data_with_missing.isna()
````

````python
data_with_missing.isna().sum()
````

### Handling missing values

Common approaches:

1. `dropna()`  
   Remove rows with missing values.

2. `fillna()`  
   Fill missing values with something, such as 0, mean, or a known estimate.

Be careful: there is no universal correct method. It depends on business meaning.


````python
data_with_missing.dropna()
````

````python
data_with_missing.fillna(0)
````

````python
# Fill Budget missing value with average budget

average_budget = data_with_missing["Budget"].mean()

data_with_missing["Budget_Filled"] = data_with_missing["Budget"].fillna(average_budget)

data_with_missing
````

## Section 16 — Mini Project 1: Project Budget and Profit Analysis

Now we combine core Pandas skills.

Scenario:

You have a project dataset.

Your tasks:

1. Create DataFrame.
2. Inspect data.
3. Create variance column.
4. Create profit column.
5. Create margin column.
6. Create budget status column.
7. Filter over-budget projects.
8. Sort by profit.
9. Group by department.
10. Export final file to CSV.


````python
projects = pd.DataFrame({
    "Project": ["Tower A", "Villa B", "Mall C", "Road D", "Office E", "Hotel F", "Warehouse G"],
    "Department": ["Construction", "Construction", "Retail", "Infrastructure", "Admin", "Hospitality", "Logistics"],
    "Budget": [100000, 150000, 200000, 175000, 125000, 300000, 180000],
    "Actual": [120000, 140000, 210000, 190000, 110000, 330000, 170000],
    "Revenue": [160000, 180000, 260000, 220000, 150000, 390000, 230000]
})

projects
````

````python
projects.head()
````

````python
projects.info()
````

````python
projects.describe()
````

````python
projects["Variance"] = projects["Actual"] - projects["Budget"]
projects["Profit"] = projects["Revenue"] - projects["Actual"]
projects["Margin"] = projects["Profit"] / projects["Revenue"]

projects
````

````python
def budget_status(row):
    if row["Actual"] > row["Budget"]:
        return "Over Budget"
    elif row["Actual"] < row["Budget"]:
        return "Under Budget"
    else:
        return "On Budget"

projects["Budget_Status"] = projects.apply(budget_status, axis=1)

projects
````

````python
over_budget = projects[projects["Budget_Status"] == "Over Budget"]

over_budget
````

````python
top_profit_projects = projects.sort_values("Profit", ascending=False)

top_profit_projects
````

````python
department_summary = projects.groupby("Department").agg({
    "Budget": "sum",
    "Actual": "sum",
    "Revenue": "sum",
    "Profit": "sum"
})

department_summary
````

## Section 18 — Pandas versus Excel

Pandas and Excel can do many similar things.

Excel is excellent for:

- quick manual inspection
- small reports
- formatting
- ad hoc calculations

Pandas is excellent for:

- repeatable analysis
- large data
- automated cleaning
- reproducible calculations
- joining many files
- preparing data for machine learning

First-principles distinction:

```text
Excel = interactive spreadsheet
Pandas = programmable spreadsheet engine
```

As a finance professional, you do not need to abandon Excel. But Pandas can help you automate repetitive Excel-style work.
