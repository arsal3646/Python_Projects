## Section 4 — Array arithmetic

This is where NumPy starts to feel powerful.

If two arrays have the same shape, NumPy can perform element-by-element operations.

Example:


````python
revenue = np.array([50000, 60000, 55000, 70000])
cost = np.array([35000, 42000, 40000, 50000])

profit = revenue - cost

print("Revenue:", revenue)
print("Cost:", cost)
print("Profit:", profit)
````

````python
margin = profit / revenue

print("Profit:", profit)
print("Revenue:", revenue)
print("Margin:", margin)
print("Margin percentage:", margin * 100)
````

## Section 5 — Aggregation: turning many numbers into one summary

Aggregation means taking many values and producing one summary value.

Common aggregate functions:

| Function | Meaning |
|---|---|
| `np.sum()` | total |
| `np.mean()` | average |
| `np.median()` | middle value |
| `np.min()` | smallest |
| `np.max()` | largest |
| `np.std()` | standard deviation |
| `np.var()` | variance |

In business, aggregation is everywhere:

- total sales
- average cost
- highest profit
- lowest margin
- volatility of monthly revenue


````python
monthly_sales = np.array([10000, 12000, 15000, 9000, 18000, 16000])

print("Total sales:", np.sum(monthly_sales))
print("Average sales:", np.mean(monthly_sales))
print("Median sales:", np.median(monthly_sales))
print("Lowest sales:", np.min(monthly_sales))
print("Highest sales:", np.max(monthly_sales))
print("Standard deviation:", np.std(monthly_sales))
print("Variance:", np.var(monthly_sales))
````

### Standard deviation — simple intuition

Standard deviation tells us how spread out values are.

If monthly sales are very similar, standard deviation is small.

If monthly sales jump up and down a lot, standard deviation is large.

Do not worry about the formula yet. For now:

```text
Small standard deviation = stable values
Large standard deviation = volatile values
```

**Volatile**  
English: changing sharply or unpredictably  
Urdu: بہت زیادہ اتار چڑھاؤ والا


````python
stable_sales = np.array([10000, 10200, 10100, 9900, 10050])
volatile_sales = np.array([5000, 20000, 8000, 25000, 7000])

print("Stable sales standard deviation:", np.std(stable_sales))
print("Volatile sales standard deviation:", np.std(volatile_sales))
````

## Section 8 — Boolean filtering

Boolean filtering means selecting values that satisfy a condition.

Example business questions:

- Which projects are over budget?
- Which months had sales above target?
- Which costs exceeded 50,000?
- Which employees received bonus above 2,000?

This is one of the most important skills for data analysis.


````python
sales = np.array([10000, 12000, 15000, 9000, 18000])

print(sales > 12000)
````

The result is an array of `True` and `False`.

```text
True  = condition satisfied
False = condition not satisfied
```

Now we use that True/False array as a filter.


````python
high_sales = sales[sales > 12000]

print(high_sales)
````

````python
costs = np.array([50000, 70000, 65000, 40000, 90000])

high_costs = costs[costs > 60000]

print(high_costs)
````

### Boolean filtering with projects

We can identify over-budget projects by comparing actuals with budgets.


````python
budgets = np.array([100000, 150000, 200000, 175000])
actuals = np.array([120000, 140000, 220000, 160000])

over_budget_filter = actuals > budgets

print("Over budget filter:", over_budget_filter)

over_budget_actuals = actuals[over_budget_filter]

print("Actual costs for over-budget projects:", over_budget_actuals)
````

## Section 9 — `np.where()`: choosing values based on conditions

`np.where()` is like an if statement for arrays.

Basic structure:

```python
np.where(condition, value_if_true, value_if_false)
```

This is useful for classification:

- Over Budget / Not Over Budget
- Profitable / Loss
- Above Target / Below Target


````python
budgets = np.array([100000, 150000, 200000])
actuals = np.array([120000, 140000, 200000])

status = np.where(actuals > budgets, "Over Budget", "Not Over Budget")

print(status)
````

### Limitation of simple `np.where()`

The above code gives only two categories:

```text
Over Budget
Not Over Budget
```

But business reality may need three categories:

```text
Over Budget
Under Budget
On Budget
```

For that, we can use nested `np.where()`.


````python
status = np.where(
    actuals > budgets,
    "Over Budget",
    np.where(actuals < budgets, "Under Budget", "On Budget")
)

print(status)
````

## Section 11 — Reshaping arrays

Reshaping changes the structure of an array without changing the data.

Example:

```text
12 monthly values
```

can become:

```text
4 quarters × 3 months
```

This is useful in time-based analysis.


````python
monthly_sales = np.array([
    10000, 12000, 15000,
    9000, 18000, 16000,
    14000, 15500, 17000,
    19000, 21000, 22000
])

print(monthly_sales)
print("Shape:", monthly_sales.shape)
````

````python
quarterly_view = monthly_sales.reshape(4, 3)

print(quarterly_view)
print("Shape:", quarterly_view.shape)
````

### Meaning of reshape(4, 3)

```text
4 rows
3 columns
```

Business interpretation:

```text
4 quarters
3 months per quarter
```


````python
# Calculate total sales per quarter

quarter_totals = np.sum(quarterly_view, axis=1)

print(quarter_totals)
````

### Understanding axis

This is very important.

For a 2D array:

```text
axis=0 means down the rows, column-wise calculation
axis=1 means across the columns, row-wise calculation
```

In the quarterly example:

```python
np.sum(quarterly_view, axis=1)
```

means:

```text
For each quarter row, add the 3 months.
```


````python
# Monthly totals by position inside quarters

monthly_position_totals = np.sum(quarterly_view, axis=0)

print(monthly_position_totals)
````

## Section 14 — Missing values with `np.nan`

Real data is messy.

Sometimes data is missing.

In NumPy, missing numerical values are often represented using:

```python
np.nan
```

`nan` means:

```text
Not a Number
```

This is very important before Pandas.


````python
sales = np.array([10000, 12000, np.nan, 15000, 9000])

print(sales)
````

````python
# Normal sum gives nan if data contains nan

print(np.sum(sales))
````

If an array contains `np.nan`, normal aggregation may return `nan`.

Use special functions:

- `np.nansum()`
- `np.nanmean()`
- `np.nanmin()`
- `np.nanmax()`


````python
print("Sum ignoring missing:", np.nansum(sales))
print("Mean ignoring missing:", np.nanmean(sales))
print("Minimum ignoring missing:", np.nanmin(sales))
print("Maximum ignoring missing:", np.nanmax(sales))
````

````python
# Detect missing values

missing_filter = np.isnan(sales)

print(missing_filter)
````

## Section 16 — Mini Project 2: Project Budget Variance Analysis

Scenario:

You manage construction projects.

For each project, you have:

- budget
- actual cost

You need to calculate:

- variance
- variance percentage
- status

This is highly relevant to finance, accounting, construction projects, and fixed assets.


````python
project_names = np.array(["Tower A", "Villa B", "Mall C", "Road D", "Office E"])

budgets = np.array([100000, 150000, 200000, 175000, 125000])
actuals = np.array([120000, 140000, 200000, 190000, 110000])

variance = actuals - budgets
variance_percentage = variance / budgets

status = np.where(
    actuals > budgets,
    "Over Budget",
    np.where(actuals < budgets, "Under Budget", "On Budget")
)

print("Project names:", project_names)
print("Budgets:", budgets)
print("Actuals:", actuals)
print("Variance:", variance)
print("Variance percentage:", variance_percentage)
print("Status:", status)
````

````python
# Print a clean project summary

for i in range(len(project_names)):
    print("Project:", project_names[i])
    print("Budget:", budgets[i])
    print("Actual:", actuals[i])
    print("Variance:", variance[i])
    print("Variance %:", round(variance_percentage[i] * 100, 2))
    print("Status:", status[i])
    print("-" * 30)
````

````python
# Overall summary

total_budget = np.sum(budgets)
total_actual = np.sum(actuals)
total_variance = total_actual - total_budget
overall_status = "Over Budget" if total_actual > total_budget else "Not Over Budget"

print("Total budget:", total_budget)
print("Total actual:", total_actual)
print("Total variance:", total_variance)
print("Overall status:", overall_status)
````

## Section 17 — What NumPy is good at and what it is not good at

NumPy is excellent for:

- numerical arrays
- fast calculations
- mathematical operations
- simulations
- vectorized operations
- foundational data science

NumPy is not ideal for:

- labelled rows and columns
- messy Excel-like data
- mixed business tables with names, dates, categories, numbers
- group-by reports by department or project type

For those, you will use Pandas.

First-principles comparison:

```text
NumPy = numerical engine
Pandas = business table analysis tool
```

A Pandas DataFrame is easier when you need column names like:

```text
Project | Department | Budget | Actual | Status
```

But NumPy is still underneath many calculations.
