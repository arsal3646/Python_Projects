# Pandas - DataFrame

```python
# Series is one-dimensional while DataFrame is two-dimensional. 
# A Series can be thought of as a single column of data, while a DataFrame is a collection of Series that share the same index.

# DataFrame is just tabular data structure with labeled axes (rows and columns). 
# It can be thought of as a dictionary of Series objects, where each Series represents a column in the DataFrame.
```

```python
# creating a DataFrame using List of Lists

import pandas as pd

student_data = [
    [100, 80, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]

students_df = pd.DataFrame(student_data, columns=["IQ", "Marks", "Salary_per_year"])

students_df.head()

print(students_df.head())
```

```python
# creating a DataFrame using Dictionaries

students_dict = {
    'iq':[100,90,120,80],
    "marks" : [80,70,100,50],
    "salary_per_year" : [10,7,14,2]
}

students_df_dict = pd.DataFrame(students_dict)
students_df_dict.head()
```
