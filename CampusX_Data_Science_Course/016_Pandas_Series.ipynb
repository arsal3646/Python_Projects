# Pandas (Series)

```python
# Why Pandas? 

# Pandas is a powerful and flexible open-source data analysis and manipulation library for Python. 

# It provides data structures like DataFrames and Series that make it easy to work with structured data. 

# Here are some reasons why Pandas is widely used:  
# 01. Easy to use.
# 02. Efficient: Pandas is built on top of NumPy - fast and efficient data processing, even with large datasets.
# 03. Versatile: Pandas can handle a wide variety of data formats, including CSV, Excel, SQL databases, etc.
# 04. Integration: Pandas integrates well with other Python libraries, such as Matplotlib, Scikit-learn.
# 05. Community support
# 06. Data Cleaning: 
#      - Pandas provides robust tools for cleaning and preparing data, including:
#           1. handling missing values
#           2. handling duplicates
#           3. handling inconsistent data types

# 07. Data Transformation: Pandas allows for easy reshaping and pivoting of data
# 08. Performance: Pandas is optimized for performance, allowing for efficient data manipulation and analysis
# 09. Visualization: Pandas has built-in support for basic data visualization
# 10. Documentation: Pandas has comprehensive documentation and tutorials
# 11. Pandas is open-source software
# 12. Data Alignment: Pandas automatically aligns data for you when performing operations on different datasets
# 13. Time Series Analysis: Pandas provides specialized functionality for working with time series data
# 14. Grouping and Aggregation: Pandas makes it easy to group data by one or more keys and perform aggregate functions
# 15. Extensibility: Pandas can be extended with custom functions and methods
```

```python
# always import pandas and numpy at the beginning of your script

import numpy as np
import pandas as pd
```

```python
# What is Series? 

# A Series is a 1D labeled array capable of holding any data type (integers, strings, float, Python objects, etc.).
# The labels, also known as the index, allow for easy access and manipulation of data.  

# We can create a Series from a list, NumPy array, or dictionary.
```

```python
# Series of strings

countries = ['Pakistan', 'India', 'Bangladesh','Afghanistan', 'Nepal', 'Sri Lanka', 'Bhutan', 'Maldives']   

pd.Series(countries)

# Series is a class inside Pandas library. 
# We can create a Series object by calling the Series class and passing in the data we want to store in the Series.
```

```python
# Series of integers

runs_Kohli = [183, 160, 122, 97, 82, 70, 65, 55]

pd.Series(runs_Kohli)
```

```python
# Custom index instead of 0,1,2,3,4,5,6,7

marks_Arsalan = [85, 90, 78, 92, 88]
subjects = ['Math', 'Science', 'English', 'History', 'Geography']

pd.Series(marks_Arsalan, index=subjects)
```

```python
marks_series = pd.Series(marks_Arsalan, index=subjects, name='Arsalan ke Marks')

marks_series
# the name is like a title or ID for the Series. 
# It can be useful when working with multiple Series or when displaying the data in a DataFrame.
```

```python
# Series from dictionary

marks_dict = {
    'Math': 85,
    'Science' : 90,
    'English' : 78,
    'History' : 92,
    'Geography' : 88
}

marks_dict
```

```python
# csv files mean Comma Separated Values files.
# In such files, data is stored in a tabular format.
# each row represents a record and each column represents a field (attribute).

# I have already downloaded the three csv files which the teacher have used in the video. You can download them from the following links:
# https://raw.githubusercontent.com/arsal3646/Python_Projects/main/CampusX_Data_Science_Course/016_DataSets_Pandas/subs.csv

subs = pd.read_csv('https://raw.githubusercontent.com/arsal3646/Python_Projects/main/CampusX_Data_Science_Course/016_DataSets_Pandas/subs.csv')

subs
```

```python
# Note: the type of the above object is automatically a DataFrame, not a Series.

type(subs)
```

```python
# we need to convert the DataFrame into a Series.
# this is done by using Squeeze method, i.e. squeeze = True

subs_series = subs.squeeze() 

# you can also doing the same directly while reading the csv file by using the squeeze parameter.
# the code would have been like this:
# subs_series = pd.read_csv('https://raw.githubusercontent.com/arsal3646/Python_Projects/main/CampusX_Data_Science_Course/016_DataSets_Pandas/subs.csv', squeeze=True)

subs_series
```

```python
subs_series.name

# the name of the Series is automatically set to the name of the column in the DataFrame.
```

```python
type(subs_series)

# now it is changed to a Series object from a DataFrame object.
```
