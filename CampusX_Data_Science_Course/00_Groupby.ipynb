```python
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# GROUPBY PRACTICE: Split - Apply - Combine
# ---------------------------------------------------------

# Create a small dataset
# The below has four columns: Team, Player, Points, Salary

df = pd.DataFrame({
    "Team": ["Red", "Red", "Blue", "Blue", "Green", "Green", "Red"],
    "Player": ["Ali", "Bilal", "Sara", "Omar", "Zain", "Noor", "Hina"],
    "Points": [12, np.nan, 18, 22, 10, 30, 16],
    "Salary": [5000, 7000, 6000, 8000, 5500, 9000, 6500]
})

df
```

```python

print("\nOriginal DataFrame:")
print(df)

# ---------------------------------------------------------
# 1. Group by Team and calculate mean points
# ---------------------------------------------------------

print("\n1. Average points by team:")
result_1 = df.groupby("Team")["Points"].mean()
print(result_1)

# Think:
# Red has 12, NaN, 16
# Pandas mean ignores NaN by default.
# So Red average = (12 + 16) / 2 = 14


# ---------------------------------------------------------
# 2. Group by Team and calculate total salary
# ---------------------------------------------------------

print("\n2. Total salary by team:")
result_2 = df.groupby("Team")["Salary"].sum()
print(result_2)


# ---------------------------------------------------------
# 3. Count players in each team
# ---------------------------------------------------------

print("\n3. Number of players by team:")
result_3 = df.groupby("Team")["Player"].count()
print(result_3)


# ---------------------------------------------------------
# 4. Multiple aggregations on one column
# ---------------------------------------------------------

print("\n4. Multiple calculations for Points:")
result_4 = df.groupby("Team")["Points"].agg(["mean", "sum", "count", "max", "min"])
print(result_4)


# ---------------------------------------------------------
# 5. Multiple columns with different aggregations
# ---------------------------------------------------------

print("\n5. Different calculations for different columns:")
result_5 = df.groupby("Team").agg({
    "Points": ["mean", "sum"],
    "Salary": ["mean", "sum"]
})
print(result_5)


# ---------------------------------------------------------
# 6. Group by more than one column
# ---------------------------------------------------------

df2 = pd.DataFrame({
    "Department": ["Finance", "Finance", "Finance", "IT", "IT", "IT"],
    "Gender": ["M", "F", "M", "F", "M", "F"],
    "Salary": [10000, 12000, 11000, 15000, 14000, 16000]
})

print("\nSecond DataFrame:")
print(df2)

print("\n6. Average salary by Department and Gender:")
result_6 = df2.groupby(["Department", "Gender"])["Salary"].mean()
print(result_6)


# ---------------------------------------------------------
# 7. Reset index after groupby
# ---------------------------------------------------------

print("\n7. Same result but converted back to normal DataFrame:")
result_7 = df2.groupby(["Department", "Gender"])["Salary"].mean().reset_index()
print(result_7)


# ---------------------------------------------------------
# 8. Practice challenge for you
# ---------------------------------------------------------

print("\nPractice Challenge:")
print("Try to write code to find maximum salary by Department.")
print("Hint: use df2.groupby('Department')['Salary'].max()")
```
