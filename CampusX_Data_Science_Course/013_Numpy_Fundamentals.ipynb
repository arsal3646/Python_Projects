# Why Numpy?

NumPy is built in C language because C is faster while Python is slower. On the other hand, syntax of Python is easier to learn and hence it became very famous among users (especially non-programming or developer).

Pandas, matplotlib, etc. are then built on NumPy. This made Numpy an excellency choice for Data Science field.

# Numpy Array vs Python Sequences

1. NumPy arrays have fixed size at creation, unlike Python lists (which can grow dynamically). If you change size of an ndarray it will create a new array and delete the original one.

2. NumPy array have one datatype, i.e. Homegenous, while lists are heterogeneous and can contain different datatypes. 

Note: Python lists are able to store different types of data because they store pointers (addresses) to the actual data rather than the data itself. This allows lists to be flexible and store various types of data. However, this means that they require more memory and are slower for numerical operations compared to NumPy arrays, which store data in a contiguous block of memory and are optimized for performance.

3. NumPy arrays are more compact than lists, which means they use less memory to store data. This is because NumPy arrays are stored in a contiguous block of memory, while lists store pointers to the actual data.

4. NumPy arrays are faster than lists for numerical operations because they are optimized for performance and use less memory. This is because NumPy arrays are implemented in C and use optimized algorithms for numerical operations, while lists are implemented in Python and are not optimized for performance.

5. NumPy arrays have a wide range of built-in functions and methods for performing mathematical operations, while lists do not have as many built-in functions for numerical operations. This makes NumPy arrays more convenient for scientific computing and data analysis tasks.

6. NumPy arrays support vectorized operations, which means you can perform element-wise operations on entire arrays without the need for explicit loops. This allows for more efficient and concise code when performing mathematical operations on large datasets.

7. NumPy arrays can be multi-dimensional, allowing for the creation of matrices and higher-dimensional arrays, while Python lists are typically one-dimensional (though they can contain other lists to create multi-dimensional structures, it is not as efficient or convenient as using NumPy arrays).

8. NumPy arrays have a fixed size at creation, while Python lists can grow dynamically. If you change the size of a NumPy array, it will create a new array and delete the original one, while Python lists can simply append or remove elements without creating a new list.

9. A growing number of libraries and frameworks in the scientific computing and data science ecosystem are built on top of NumPy, making it a fundamental tool for many applications in these fields. This widespread adoption has further solidified NumPy's position as a key component of the Python scientific computing stack.

```python
# Boolean data type can also be specified while creating the array.

array_bool_1 = np.array([1,0,2,0,-1], dtype=bool)      

# the value must be either 0 or 1? No, any non-zero value (e.g. 2, -1, 1000, etc.) will be considered as True and zero will be considered as False.

print(array_bool_1)
print(type(array_bool_1))
```

```python
# let us see boolean array by using True or False values only.

array_bool_2 = np.array([True, False, True, False, True], dtype=bool)
print(array_bool_2)
print(type(array_bool_2))
```

```python
# let us see boolean array by using values other than True or False or 1 or 1

array_bool_3 = np.array(['Arsalan', 5, "e", -1, True, 0, 0.0, False,"", None], dtype=bool)
print(array_bool_3)
print(type(array_bool_3))

# Conclusion: Python converts values to boolean using truthiness rules.

# Values considered False:

# 1. 0
# 2. 0.0
# 3. False
# 4. Empty string ""
# 5. Empty containers such as [], (), {}, set()
# 6. None

# Most other values are considered True.
```

```python
# [],{},set(),(). - these cannot be tested for boolean values as they are empty containers, not scalar values.

# [], {}, set(), and () are collections, not simple scalar values, so NumPy gets confused when forcing everything into one clean boolean array.

bool_list = bool([]) # this to see whether this container (list) is True or False. It will return False as it is an empty container.

print(bool_list)
print(type(bool_list))
```

```python
bool_dict = bool({}) # this to see whether this container (dictionary) is True or False. It will return False as it is an empty container.
print(bool_dict)
print(type(bool_dict))
```

```python
bool_set = bool(set()) # this to see whether this container (set) is True or False. It will return False as it is an empty container.
print(bool_set)
print(type(bool_set))
```

```python
bool_tuple = bool(()) # this to see whether this container (tuple) is True or False. It will return False as it is an empty container.
print(bool_tuple)
print(type(bool_tuple))
```

```python
# range(0) also returns False as it is an empty container.

print(bool(range(0)))
print(type(bool(range(0))))
```

# Summary: 

In Python, empty containers (list, dictionary, set(), range(), and tuple) are considered False when converted to boolean values.

Other than empty containers, we saw 0, 0.0, False, empty string "", and None are also considered False when converted to boolean values.

Almost everything else is considered True when converted to boolean values including 1, -1, 2, -2, 1000, -1000, "Arsalan", "e", " ", etc. 

The last one is a space character, it is a non-empty string, which is also considered True when converted to boolean values.

```python
# now we study np.arange() function
# It isused to create an array of evenly spaced values within a given interval. 
# It is similar to the built-in range() function but returns an array instead of a list.

array_arange1 = np.arange(100,110)
print(array_arange1)
print(type(array_arange1))

# Very important note: the last value (stop value) is not included in the array, it is exclusive. 
# So the array will include values from 100 to 109 only, not 110.
# This is same as the built-in range() function in Python, where the stop value is also exclusive.

array_arange2 = np.arange(50,55)    # this will print 50, 51, 52, 53, and 54 only, not 55.
print(array_arange2)
print(type(array_arange2))
```

```python
# STEPS in np.arange() function
# In other words, strides is the step size between values in the array. 
# It is an optional argument that can be specified to create an array with a specific step size.

# np.arange(start, stop, step)

# e.g. print even number between 0 and 10 by using np.arange()

np.arange(0, 10, 2) # this will print 0, 2, 4, 6, and 8 only, not 10. The step size is 2.
```

```python
# print even numbers from 0 to infinity by using np.arange(). 
# However, this will not work as it will end up in an infinite loop and will consume all the memory of the computer. 
# So we should avoid using np.arange() with infinity as stop value.  

# np.arange(0, np.inf, 2) # this will print even numbers from 0 to infinity. The step size is 2. 

# The stop value is infinity, so it will keep generating even numbers indefinitely.
```

```python
# print table of 5 by using np.arange()

np.arange(0, 51, 5) # this will print 0, 5, 10, 15, 20, 25, 30, 35, 40, and 45 only, not 50. The step size is 5.
```

```python
# reshape() function is used to change the shape of an array without changing its data.
# It takes a tuple as an argument which specifies the new shape of the array.

array_reshaped = np.arange(100,2000,200).reshape(10,1) 

# this will print 100, 200, 300, ..., 1900 only, not 2000. The step size is 200.
# the reshape() function will change the shape of the array to 10 rows and 1 column.

print(array_reshaped)
print(type(array_reshaped))
```
