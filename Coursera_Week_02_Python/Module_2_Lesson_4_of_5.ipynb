## Video : Regular Functions

```python
class Book:     
    def __init__(self, title, author, pages):
        self.title = title
        self.pages = pages
        self.author = author
```

#### Function parameters

```python
def calculate_price(price, discount):
    return price - (price * discount)

#final_price = calculate_price(100, 0.1)
#print(f"Final price: ${final_price}")  

print(f"Final price: ${calculate_price(100, 0.1)}")  
```

#### Default Parameter Values

```python
def create_book(title, author, pages=200):
    return Book(title, author, pages)

book1 = create_book("Python Basics", "John Doe")
book2 = create_book("Advanced Python", "Jane Smith", 350)
```

#### Return Statement

```python
def get_book_info(book):
    return book.title, book.author, book.pages

title, author, pages = get_book_info(book1)
print(title, author, pages)
```

#### Keyword Arguments

```python
def describe_book(title, author, pages):
    return f"{title} by {author}, with {pages} pages"

description = describe_book(pages=340, author="John Doe", title="Python Mastery")
print(description)
```

#### Functions with Multiple Returns

```python
def analyze_book(book):
    if book.pages > 300:
        return "Long book", book.pages
    else:
        return "Short book", book.pages

category, page_count = analyze_book(book1)
print(category)
```

## Video : Lambda Functions

#### Simple Lambda Function

```python
# Regular function
def square(x):
    return x ** 2

# Equivalent lambda function
square_lambda = lambda x: x ** 2

print(square(5))       
print(square_lambda(5))  
```

#### Lambda Functions with Multiple Arguments

```python
sum = lambda a, b: a + b
print(sum(5, 3))   

product = lambda x, y, z: x * y * z
print(product(2, 3, 4))   
```

#### Lambda Functions in Book Operations

```python
books = [
    Book("Python Basics", "John Doe", 200),
    Book("Data Science", "Jane Smith", 350),
    Book("Machine Learning", "Bob Johnson", 400)
]

# Sort books by number of pages
sorted_books = sorted(books, key=lambda x: x.pages)
```

#### Lambda with map()

```python
book_titles = list(map(lambda book: book.title, books))
print(book_titles)  
```

#### Lambda with filter()

```python
long_books = list(filter(lambda book: book.pages > 300, books))
for book in long_books:
    print(f"{book.title}: {book.pages} pages")
```
