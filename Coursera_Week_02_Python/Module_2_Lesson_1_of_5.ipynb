## Video: Input and Print

#### The input() Function

```python
name = input("Enter your name: ")
print("Hello, " + name + "!")
```

#### Type Conversion with input()

```python
age = int(input("Enter your age: "))
print("Next year, you'll be", age + 1)
```

#### The print() Function

```python
print("Hello, World!")
x = 5
print("The value of x is:", x)
```

#### Formatting Output with print()

```python
name = "Alice"
age = 30
print(f"{name} is {age} years old.")
```

#### Printing Multiple Lines

```python
print("""
This is a
multi-line
output.
""")
```

#### Combining input() and print()

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
print(f"In 10 years, {name} will be {age + 10} years old.")
```

## Video : Intendentation

#### If Statement Indentation

```python
x = 10
if x > 5:
    print("x is greater than 5")
    print("This is still part of the if block")
print("This is outside the if block")
```

#### Indentation Error in if statement

```python
x = 10
if x > 5:
print("x is greater than 5")
print("This is still part of the if block")
print("This is outside the if block")
```

#### Example: Function Indentation

```python
def greet(name):
    print("Hello,", name)
    print("Welcome to Python!")

greet("Alice")
```

#### Indentation Error in function

```python
def greet(name):
    print("Hello,", name)
print("Welcome to Python!")

greet("Alice")
```

#### Example: Loop Indentation

```python
for i in range(3):
    print("Iteration", i)
    if i == 1:
        print("Middle iteration")
print("Loop finished")
```

#### Indentation Error in Loop Indentation

```python
for i in range(3):
print("Iteration", i)
    if i == 1:
        print("Middle iteration")
print("Loop finished")
```
