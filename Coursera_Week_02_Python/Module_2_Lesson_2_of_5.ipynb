#### Property decorators

```python
class Book:
    def __init__(self, title, price):
        self._title = title
        self._price = price

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        if value < 0:
            raise ValueError("Price cannot be negative")
        self._price = value
```

```python
book = Book("Data Analytics Basics", 29.99)
print(book.price)
```

```python
book.price = -34.99
print(book.price)
```

## Video : Methods

#### Instance Methods

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def get_info(self):
        return f"{self.title} by {self.author}, {self.pages} pages"

book = Book("Data Science Handbook", "Jake VanderPlas", 500)
print(book.get_info())
```

#### Class Methods

```python
class Book:
    num_books = 0 # class attribute
    def __init__(self, title, author):
        self.title = title
        self.author = author
        Book.num_books += 1
    @classmethod # class methods
    def from_string(cls, book_str):
        title, author = book_str.split(' by ')
        return cls(title, author)
    @classmethod
    def get_total_books(cls):
        return cls.num_books

book = Book.from_string("Clean Code by Robert Martin")
print(Book.get_total_books())  
```

#### Method Overriding

```python
# superclass
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        def __str__(self):
            return f"{self.title} by {self.author}"

#subclass
class Ebook(Book):
    def __init__(self, title, author, format):
        super().__init__(title, author)
        self.format = format
    
    def __str__(self):
        return f"{super().__str__()} ({self.format})"


book = Book("Clean Code", "martin")
print(book.title)
ebook = Ebook("Python Basics", "Real Python", "PDF")
print(ebook.title)
```

#### Properties

```python
class Book:
    def __init__(self, title, price):
        self._title = title
        self._price = price
    @property
    def price(self):
        return self._price
    @price.setter
    def price(self, value):
        if value < 0:
            raise ValueError("Price cannot be negative")
        self._price = value

book = Book("Data Analytics", 29.99)
print(book.price)  
book.price = 34.99
print(book.price)  
```

## Video : Operators

#### Arithmetic and Comparison operator overloading

```python
class Book:
    def __init__(self, title, price, pages):
        self.title = title
        self.price = price
        self.pages = pages
    
    def __add__(self, other):
        return self.price + other.price
    
    def __lt__(self, other):
        return self.pages < other.pages
    
    def __eq__(self, other):
        return self.title == other.title and self.pages == other.pages

```

```python
book1 = Book("Python Basics", 29.99, 544)
book2 = Book("Advanced Python", 39.99, 876)
total_price = book1 + book2
print(f"Total price: ${total_price:.2f}")  
print(book1 < book2) 
print(book1 == Book("Python Crash Course", 34.99, 544)) 
```

#### Logical operators

```python
class Book:
    def __init__(self, title, in_stock, price):
        self.title = title
        self.in_stock = in_stock
        self.price = price

book = Book("Data Science Handbook", True, 49.99)
if book.in_stock and book.price < 50:
    print(f"{book.title} is available and affordable!")
```

#### Assignment Operator  overloading

```python
class Book:
    def __init__(self, title, quantity):
        self.title = title
        self.quantity = quantity
    
    def __iadd__(self, other):
        self.quantity += other
        return self

book = Book("Python for Data Analysis", 5)
book += 3
print(f"Updated quantity: {book.quantity}") 
```

#### Bitwise Operator overloading

```python
class Book:
    def __init__(self, title, categories):
        self.title = title
        self.categories = categories
    
    def __and__(self, other):
        return set(self.categories) & set(other.categories)

book1 = Book("Python Machine Learning", ["Python", "AI", "Data Science"])
book2 = Book("Deep Learning with Python", ["Python", "AI", "Neural Networks"])
common_categories = book1 & book2
print(f"Common categories: {common_categories}")
```

#### Membership Operator overloading

```python
class BookCollection:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def __contains__(self, book_title):
        return any(book.title == book_title for book in self.books)

collection = BookCollection()
collection.add_book(Book("Python Cookbook", 574))
print("Python Cookbook" in collection)
```
