## Video : Reading Files

#### Reading files

```python
# Purpose: Learning basic python file handling.

with open('books.txt', 'r') as f:   
    content = f.read()                  
    print(content)


# The 'with' statement is used to open the file and it will automatically close the file after the block of code is executed.

#open function()
# open() function is used to open the file
# open() takes two arguments:
#   1. file name, e.g. 'books.txt'
#   2. the mode in which we want to open the file, e.g. 
#           'r' for read. 
#           'w' for write 
#           'a' for append

    
# # read - r
# f.read() reads the entire content of the file and returns it as a string.
# if the file is not found, it will raise a FileNotFoundError.
# the returned data can be stored in a variable, e.g. content = f.read()
# Once the above line is executed, the variable 'content' will contain the entire content of the file as a string.


# write - w




# append - a
    


```

#### reading line by line

```python
with open('books.txt', 'r') as file:
    for line in file:
        print(line.strip())  
```

#### Working with CSV Files

```python
import csv

with open('books.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        title, author, pages = row
        print(f"Book: {title} by {author}, Pages: {pages}")
```

#### Reading from JSON Files

```python
import json

with open('book_info.json', 'r') as file:
    data = json.load(file)
print(data)
```

## Video : Writing to Files

#### Writing Text to a File

```python
with open('books.txt', 'w') as file:
    file.write("Python Basics: A Practical Introduction\n")
    file.write("Data Science Essentials in Python\n")
    
# w - write - rewrite the contents of file
# a - append - add the new to the end of the file
```

#### Writing Multiple Lines at Once

```python
books = ["Machine Learning Guide", "Advanced Python Programming", "SQL for Data Analysis"]
with open('books.txt', 'a') as file:
    file.writelines(book + '\n' for book in books)
```

```python
with open('books.txt', 'r') as file:
    content = file.read()
    print(content)
```

#### Writing Structured Data (CSV)

```python
import csv

books = [
    ['Title', 'Author', 'Year'],
    ['Python Crash Course', 'Eric Matthes', '2019'],
    ['Data Science from Scratch', 'Joel Grus', '2019']
]

with open('books.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(books)
```

#### Writing JSON Data

```python
import json

book_data = {
    'title': 'Python for Data Analysis',
    'author': 'Wes McKinney',
    'year': 2022,
    'topics': ['pandas', 'numpy', 'matplotlib']
}

with open('book_info.json', 'w') as file:
    json.dump(book_data, file, indent=4)
```
