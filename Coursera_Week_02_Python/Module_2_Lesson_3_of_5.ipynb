## Video : Conditional Statement

```python
class Book:
    category = "Programming"      
    def __init__(self, title, pages, price, format):
        self.title = title
        self.pages = pages
        self.price = price
        self.format = format
```

#### if

```python
book = Book("Python Basics", 50, 39.99, 'digital')
if book.pages > 100:
    print(f"{book.title} is a substantial book.")
```

#### else

```python
if book.pages > 300:
    print(f"{book.title} is a long book.")
else:
    print(f"{book.title} is a relatively short book.")
```

#### elif

```python
book = Book("Python Basics", 500, 39.99, 'digital')
if book.pages < 100:
    print("Short read")
elif book.pages < 300:
    print("Medium length")
else:
    print("Long read")
```

#### combining conditions

```python
if book.pages > 200 and book.price < 30:
    print(f"{book.title} is a good value for its length.")
```

#### Ternary Operator

```python
book_type = "E-book" if book.format == "digital" else "Physical book"
print(book_type)
```

#### Nested Conditional Statements

```python
if book.category == "Programming":
    if book.pages > 500:
        print("Comprehensive programming guide")
    else:
        print("Introductory programming book")
else:
    print("Non-programming book")
```

#### Early return

```python
def categorize_book(book):
    if book.pages < 100:
        return "Short read"
    if book.pages < 300:
        return "Medium length"
    return "Long read"

categorize_book(book)
```

## Video : Nested Conditional Statement

#### Two levels of nesting

```python
def categorize_book(book):
    if book.category == "Programming":
        if book.pages > 500:
            return "Comprehensive programming guide"
        else:
            return "Introductory programming book"
    else:
        if book.pages > 300:
            return "Long non-programming book"
        else:
            return "Short non-programming book"
```

#### Multiple Levels of Nesting

```python
def analyze_book(book):
    if book.category == "Programming":
        if book.pages > 500:
            if book.price > 50:
                return "Premium comprehensive guide"
            else:
                return "Affordable comprehensive guide"
        else:
            return "Introductory programming book"
    else:
        return "Non-programming book"
```

#### Using elif in Nested Structures

```python
def classify_book(book):
    if book.category == "Fiction":
        if book.pages < 200:
            return "Short story"
        elif book.pages < 400:
            return "Novel"
        else:
            return "Epic"
    elif book.category == "Non-fiction":
        if book.pages < 300:
            return "Brief non-fiction"
        else:
            return "Comprehensive non-fiction"
    else:
        return "Uncategorized"
```

#### Combining Nested Conditionals with Logical Operators

```python
def is_bestseller(book):
    if book.price > 20.0:
        if book.pages > 300 or (book.pages > 200 and book.format == 'digital'):
            return True
    return False

is_bestseller(book)
```

#### Early Return

```python
def categorize_book(book):
    if book.category != "Programming":
        return "Non-programming book"
    if book.pages <= 500:
        return "Introductory programming book"
    return "Comprehensive programming guide"

categorize_book(book)
```

## Video : For and While loop

```python
class Book:
    category = "Programming"      
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
```

#### for loop

```python
books = [Book("Python Basics", 200), Book("Data Science", 350), Book("ML Intro", 280)]
for book in books:
    print(f"{book.title}: {book.pages} pages")
```

#### for loop in range()

```python
for i in range(5):
    print(i)  
```

```python
for i in range(1, 6): # start, end
    print(i)  
```

#### For Loops: Enumerate

```python
for index, book in enumerate(books, start=1):
    print(f"{index}. {book.title}")
```

#### While Loop Example: Reading Books

```python
total_pages = 1000
pages_read = 0
while pages_read < total_pages:
    pages_read += 50
    print(f"You've read {pages_read} pages")
print("You've finished the book!")
```

#### Infinite Loops and Break Statement

```python
while True:
    user_input = input("Enter a book title (or 'quit' to exit): ")
    if user_input.lower() == 'quit':
        break
    print(f"You entered: {user_input}")
```

#### Continue Statement

```python
for book in books:
    if book.pages <= 200:
        continue
    print(f"{book.title} is a lengthy book")
```

#### Nested Loops

```python
categories = ["Fiction", "Non-fiction"]
lengths = ["Short", "Medium", "Long"]
for category in categories:
    for length in lengths:
        print(f"{length} {category} book")
```
